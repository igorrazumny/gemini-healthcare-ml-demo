# 🧠 Gemini API Example: Machine Learning in Healthcare
This notebook demonstrates how to use the `google-generativeai` Python SDK to interact with Gemini 1.5 Flash for summarizing healthcare use cases.

You can use this notebook as a starting point for your GitHub ML portfolio.

```python
# ✅ Import and configure Gemini
from google import generativeai as genai

# Replace with your own API key
genai.configure(api_key="your-api-key-here")
```

```python
# ✅ Create model and send a prompt
model = genai.GenerativeModel("gemini-1.5-flash")

response = model.generate_content(
    "Summarize the role of machine learning in healthcare."
)

print(response.text)
```
